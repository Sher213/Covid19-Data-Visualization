```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
%matplotlib inline

covidData = pd.read_csv("covid_global.csv")
ctrys = gpd.read_file('worldcountries/99bfd9e7-bb42-4728-87b5-07f8c8ac631c2020328-1-1vef4ev.lu5nk.shp')
merged = ctrys.set_index('CNTRY_NAME').join(covidData.set_index('Country/Region'))
merged = merged.reset_index()
merged = merged.fillna(0)
linewidth=3
fig,ax = plt.subplots(1, figsize=(40,20))
ax.axis('off')
ax.set_title('Total Confirmed Covid Cases (Millions)', fontdict={'fontsize': '40', 'fontweight' : '3'})
color = 'Greens'
vmin, vmax = 0, 4000000
sm = plt.cm.ScalarMappable(cmap=color, norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []
cbar = fig.colorbar(sm)
cbar.ax.tick_params(labelsize=20)
merged.plot('Confirmed', cmap=color, linewidth=linewidth, ax=ax, edgecolor='0.8', figsize=(40,20))
plt.show()
```

```python
fig,ax = plt.subplots(1, figsize=(40,20))
ax.axis('off')
ax.set_title('Total Recoveries', fontdict={'fontsize': '40', 'fontweight' : '3'})
color = 'Blues'
vmin, vmax = 0, 1900000
sm = plt.cm.ScalarMappable(cmap=color, norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []
cbar = fig.colorbar(sm)
cbar.ax.tick_params(labelsize=20)
merged.plot('Recovered', cmap=color, linewidth=linewidth, ax=ax, edgecolor='0.8', figsize=(40,20))
plt.show()
```
